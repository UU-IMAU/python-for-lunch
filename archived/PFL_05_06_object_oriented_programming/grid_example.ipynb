```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from grid import Grid, RectilinearGrid, CurvilinearGrid
```

```python
recti_grid = RectilinearGrid(lon=np.linspace(-180, 180, 19),
                             lat=np.linspace(-70, 70, 15),
                             time=np.arange(0,4*86400+1,86400))

recti_grid.print_whoAmI()
```

```python
r = np.arange(1, 10, 1)
theta = np.pi / 180. * np.arange(30, 151, 15)
(r_g, theta_g) = np.meshgrid(r, theta)

lon = r_g * np.cos(theta_g)
lat = r_g * np.sin(theta_g)


curvi_grid = CurvilinearGrid(lon=lon,
                            lat=lat,
                            time=np.arange(0,4*86400+1,86400))

curvi_grid.print_whoAmI()
```

```python
plt = recti_grid.plot()
plt.show();
```

```python
plt = curvi_grid.plot()
plt.show();
```
