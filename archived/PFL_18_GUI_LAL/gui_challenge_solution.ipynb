# Working with graphical user interfaces

## Challenge

### Level 1
Create a tkinter graphical user interface (GUI) with 5 elements:

- a Label with text "Counter"
- a Label with text to be modifed (level 3)
- a Button to increment a counter
- a Button to decrement a counter
- a Button to reset a counter to 0

### Level 2
Define a global variable 'value' that can be modifeid with the buttons

Add functions that update 'value'  

Call functions from the buttons (e.g. "command = lambda: val_increment()")

### Level 3

Add a tk.StringVar that is set by the functions and used to update the label in one of the GUI elements




## Solution L1


```python
import tkinter as tk;

# Set up the root GUI
root = tk.Tk()
root.geometry("200x200") 
root.title("Test GUI") 
root.attributes("-topmost", True)

# Add a static label
ls = tk.Label(root, text="Counter", height=2)
ls.pack(fill=tk.BOTH, expand=False)

# Add an output Label that responds to other GUI elements
lo = tk.Label(root, text="0", height=2)
lo.pack(fill=tk.BOTH, expand=False)

# Add a incr button
bi = tk.Button(root, text="Increment", height=2)
bi.pack(fill=tk.BOTH, expand=False)

# Add a decr button
bd = tk.Button(root, text="Decrement", height=2)
bd.pack(fill=tk.BOTH, expand=False)

# Add a reset button
br = tk.Button(root, text="Reset", height=2)
br.pack(fill=tk.BOTH, expand=False)

# Run the main tool
root.mainloop()
```

## Solution L3

```python
import tkinter as tk;

# Set up the root GUI
root = tk.Tk()
root.geometry("200x200") 
root.title("Test GUI") 
root.attributes("-topmost", True)

# create a value to be modified
value = 0
text_var = tk.StringVar(root)
text_var.set(str(value))

# Add a static label
ls = tk.Label(root, text="Counter", height=2)
ls.pack(fill=tk.BOTH, expand=False)

# Add an output Label that responds to other GUI elements
lo = tk.Label(root, textvariable=text_var, height=2)
lo.pack(fill=tk.BOTH, expand=False)

# Add a incr button
bi = tk.Button(root, text="Increment", height=2, command = lambda: val_increment())
bi.pack(fill=tk.BOTH, expand=False)

# Add a decr button
bd = tk.Button(root, text="Decrement", height=2, command = lambda: val_decrement())
bd.pack(fill=tk.BOTH, expand=False)

# Add a reset button
br = tk.Button(root, text="Reset", height=2, command = lambda: val_reset())
br.pack(fill=tk.BOTH, expand=False)

# Some actions to be called from the buttons
def val_increment():
    global value
    global text_var
    value = value + 1;
    text_var.set(str(value))

def val_decrement():
    global value
    global text_var
    value = value - 1;
    text_var.set(str(value))
    
def val_reset():
    global value
    global text_var
    value = 0;
    text_var.set(str(value))


# Run the main tool
root.mainloop()

# Returns value when GUI is closed!
print(value)
```
