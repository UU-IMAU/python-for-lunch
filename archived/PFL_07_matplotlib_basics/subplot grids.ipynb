This example demonstrates how multiple axes can be created quickly using `subplot2grid` without thinking too much about their relative coordinates. 

https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot2grid.html

```python
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.patches as patches
import numpy as np

%matplotlib inline
```

```python
"""
EXAMPLE MODIFIED FROM: https://github.com/rabernat/python_teaching/blob/master/one_day_workshop/02_numpy_and_matplotlib.ipynb
"""

# two dimensional grids
x = np.linspace(-2*np.pi, np.pi, 100)
y = np.linspace(-np.pi, np.pi, 50)
xx, yy = np.meshgrid(x, y)
f = np.sin(xx) * np.cos(0.5*yy)


fig = plt.figure(figsize=(10,7))
ax1 = plt.subplot2grid((7,6),(0,1),colspan=5)
ax2 = plt.subplot2grid((7,6),(1,0),rowspan=5)
ax3 = plt.subplot2grid((7,6),(1,1),rowspan=5, colspan=5)
ax4 = plt.subplot2grid((7,6),(6,1),colspan=5)

ax1.plot(x, f.mean(axis=0))
ax2.plot(f.mean(axis=1), y)
cn = ax3.pcolormesh(x, y,  f)

plt.colorbar(cn, cax=ax4, orientation='horizontal')


plt.tight_layout()
```
